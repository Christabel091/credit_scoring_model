```python
# Plotting the data
import matplotlib.pyplot as plt
plt.plot(df['Monthly_Inhand_Salary'], df['Credit_Score'], label='y=Credit Score of customers over their Age ', marker='o')  # Plotting x vs y with a line and markers
plt.xlabel('X-axis')  # Labeling the x-axis
plt.ylabel('Y-axis')  # Labeling the y-axis
plt.title('Straight Line Plot')  # Adding a title
plt.legend()  # Adding a legend
plt.grid(True)  # Adding a grid
plt.show()  # Displaying the plot
```
