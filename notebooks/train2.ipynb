```python
# Example: Label Encoding for a single column
# Replace '_' with a specific value for missing data (if applicable)
df['Payment_Behaviour'] = df['Payment_Behaviour'].fillna('Unknown')  # Example for handling missing values
label_encoder = LabelEncoder()
df['Payment_Behaviour_LabelEncoded'] = label_encoder.fit_transform(df['Payment_Behaviour'])

# Handle NaN values in 'Type_of_Loan' column and convert to string
df['Type_of_Loan'] = df['Type_of_Loan'].fillna('').astype(str)

# Split the string values into separate columns
df['Type_of_Loan'] = df['Type_of_Loan'].str.split(',')

# Flatten the list of lists into a single list of unique categories
unique_types = set([item.strip() for sublist in df['Type_of_Loan'] for item in sublist if item])

# Create new columns for each unique category and fill with 0s and 1s
for typ in unique_types:
    df[typ] = df['Type_of_Loan'].apply(lambda x: 1 if typ in x else 0)

# Drop the original 'Type_of_Loan' column
df = df.drop('Type_of_Loan', axis=1)

# Now df contains encoded columns for each type of loan
print(df.head())
```
